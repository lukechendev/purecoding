```python
from sklearn.neighbors import KNeighborsClassifier

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('/home/lchen154/Downloads/AI/Kannada-MNIST/train.csv')
test = pd.read_csv('/home/lchen154/Downloads/AI/Kannada-MNIST/test.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
train_set = train.drop('label', axis = 1)
train_label = train.label
```

```python
train_set.head()
```

```python
train_label.head()
```

```python
train.shape
```

```python
train_image = train_set.iloc[0:500].values
train_labels = train_label.iloc[0:500].values
```

```python
classifier = KNeighborsClassifier(n_neighbors = 3)
```

```python
// training
classifier.fit(train_image, train_labels)
```

```python
test_set = test.drop('id', axis = 1)
test_set = test_set.iloc[0:500].values
test_set.shape
```

```python
print(train_set.shape)
print(test_set.shape)
```

```python
prediction = classifier.predict(test_set)
```

```python
plt.figure()
for i in range (4):
    plt.subplot(1, 4, i+1)
    img = test_set[i].reshape(28, 28)
    plt.imshow(img)
    plt.title('Predicted: ' + str(prediction[i]))
    plt.axis('off')
plt.show()
```
